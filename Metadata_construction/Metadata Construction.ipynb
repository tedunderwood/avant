```python
retrospective['is_retroavant'] = True
retrospective = retrospective.rename(columns={"Vol_Id": "docid", "First_Pub": "retrospective_firstpub"})

# manually change the docid from volumelist to hathilist
retrospective.loc[1, 'docid'] = 'mdp.39015056245924'
retrospective.loc[11, 'docid'] = 'mdp.39015008982822'
retrospective.loc[16, 'docid'] = 'mdp.39015000618473'
retrospective.loc[18, 'docid'] = 'mdp.39015009100812'
retrospective.loc[19, 'docid'] = 'uc1.32106010568639'
retrospective.loc[23, 'docid'] = 'inu.30000007661410'
retrospective.loc[25, 'docid'] = 'uc1.32106005857609'
retrospective.loc[30, 'docid'] = 'uc1.$b242632'
retrospective.loc[32, 'docid'] = 'mdp.39015050713448'

print(len(retrospective))
retrospective.head()
```

```python
topicmeta.columns
```

```python
hathi_to_merge = hathimeta[['docid', 'author', 'title', 'latestcomp']]
topic_to_merge = topicmeta[['docid', 'in_topic_model', 'hathi_author', 'hathi_title', 'topicmodel_firstpub', 'birthyear', 'us_national', 'authof3ormore']]
merged_meta = hathi_to_merge.merge(topic_to_merge, 'outer', on = 'docid') # this is set as outer as we want works either in hathi or topic
# but in fact there is no difference: all docids in topicmeta are in hathimeta
merged_meta = merged_meta.merge(brd[['docid','in_BRD', 'brd_review_year', 'rownum']], 'left', on = 'docid')
merged_meta = merged_meta.merge(bestsellers[['docid','is_bestseller', 'bestseller_year']], 'left', on = 'docid')
merged_meta = merged_meta.merge(prize[['docid','is_prize_author']], 'left', on = 'docid')
merged_meta = merged_meta.merge(retrospective[['docid','is_retroavant', 'retrospective_firstpub']], 'left', on = 'docid')
merged_meta['consensus_earliest_date'] = merged_meta[['latestcomp', 'topicmodel_firstpub', 'brd_review_year',
                                                     'bestseller_year', 'retrospective_firstpub']].min(axis = 1)
merged_meta['author_age'] = merged_meta['consensus_earliest_date'] - merged_meta['birthyear']
merged_meta
```

```python
# sanity check: in_topic_model has 29351 instances, in_BRD 8294 instances, is_bestseller 778 instances, is_prize_author 700 instances
# these are all the same as the size of original topicmeta/BRD match/bestseller match/prize author match
# however, there are 39 instances in is_retroavant, while retro avant match has 48 instances: 9 are missing
merged_meta.count()
```

```python
# rename nan to False
merged_meta['in_topic_model'] = merged_meta['in_topic_model'].replace(np.nan, False)
merged_meta['in_BRD'] = merged_meta['in_BRD'].replace(np.nan, False)
merged_meta['is_bestseller'] = merged_meta['is_bestseller'].replace(np.nan, False)
merged_meta['is_prize_author'] = merged_meta['is_prize_author'].replace(np.nan, False)
merged_meta['is_retroavant'] = merged_meta['is_retroavant'].replace(np.nan, False)
merged_meta
```
